```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

xticks = range(1,11)

# assuming a string 
# 250/250 [==============================] - 24s 32ms/step - loss: 0.2817 - accuracy: 0.9112 - val_loss: 0.0035 - val_accuracy: 0.9990

def read_times(filename):
    result = []
    with open(filename) as f:
        for _ in xticks:
            f.readline()  # epoch line
            temp = f.readline()  # Result line
            temp = temp.split(":")[1]
            # print(temp)
            temp = temp.split("accuracy")[0]
            # print(temp)
            temp = temp.split(" -")[0]
            # print(temp)
            result.append(float(temp))
    return result

def read_val(filename):
    result = []
    with open(filename) as f:
        for _ in xticks:
            f.readline()  # epoch line
            temp = f.readline()  # Result line
            temp = temp.split(":")[3]
            # print(temp)
            temp = temp.split("accuracy")[0]
            # print(temp)
            temp = temp.split(" -")[0]
            # print(temp)
            result.append(float(temp))
    return result

loss = read_times(f"nn.out")
plt.plot(xticks, loss, "-")
val_loss = read_val(f"nn.out")
plt.plot(xticks, val_loss, "-")


plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss vs Epoch for CNN trained on Original Data")
plt.legend(['training_loss', 'validation_loss'])
plt.xticks(xticks)
plt.savefig('loss.png')
```
